```python
import os
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt

from keras.utils import np_utils
```

```python
SISI_DIR = '/media/ben/data/kaggle/SISI'
```

***load and prepare the data***

```python
# load datset
f = open(os.path.join(SISI_DIR,'shipsnet.json'))
dataset = json.load(f)
f.close()
```

```python
# assign Xy 
X = np.array(dataset['data']).astype('uint8')
y = np.array(dataset['labels']).astype('uint8')
```

***Image Normalisation: [How](https://machinelearningmastery.com/how-to-manually-scale-image-pixel-data-for-deep-learning/) and [Why](https://medium.com/@urvashilluniya/why-data-normalization-is-necessary-for-machine-learning-models-681b65a05029)***
<br>
Pixel values are often unsigned integers ranging from between 0 and 255. This can present issues later on as machine leanring models typically expect values within a common range (e.g. 0-1). To overcome scaling issues we can normalize each pixel value to the between 0-1.

```python
X = X / 255
```

```python
# prepare X data
# The dataset contains 4000 images 
# each image is represented as a vector of length 19200 elements.
# convert these to an 80*80*3 (RGB band) image.
n_spectrum = 3 # color chanel (RGB)
weight = 80
height = 80
X = X.reshape([-1, n_spectrum, weight, height]).transpose([0,2,3,1])
X.shape
```

```python
# prepare y data
y = np_utils.to_categorical(y, 2)
```

```python
# plot some examples
fig,axs = plt.subplots(4,5,figsize=(9,9),sharey=False,sharex=False)
nbins = 50
for e,i in enumerate([0,1,-1,-2]):
    # plot colour img
    axs[e,0].imshow(X[i]) 
    
    for band,c in zip(range(3),['r','g','b']):
        # plot each RGB colour band
        axs[e,band+1].imshow(X[i][:,:,band])
        
        # plot a histogram of RGB intensity
        axs[e,4].hist(X[i][:,:,band].flatten(),bins=nbins,color=c,alpha=0.4)

# axis styles
for ax in axs.flatten():
    ax.set_yticklabels([''])
    ax.set_xticklabels([''])
        
axs[0,0].set_ylabel('Ship')
axs[1,0].set_ylabel('Ship')
axs[2,0].set_ylabel('Not Ship')
axs[3,0].set_ylabel('Not Ship')

axs[0,0].set_title('Colour')
axs[0,1].set_title('Red')
axs[0,2].set_title('Green')
axs[0,3].set_title('Blue')
axs[0,4].set_title('RGB Intensity');
```

# Modelling

#### Train & Test splits

```python
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.metrics import confusion_matrix
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.3 ,random_state = 8623)

print(f'Train : {len(x_train)}\nTest : {len(x_test)}')
```

### Model 1 - traditional ML models

```python
# import models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
```

```python
def TrainTest_ML_Models(x_train,y_train,x_test,y_test):     
    
    # define models to test
    models = []
    models.append(('LR', LogisticRegression(solver='lbfgs')))
    models.append(('RF', RandomForestClassifier(n_estimators=10)))
    models.append(('KNN', KNeighborsClassifier()))
    models.append(('SVM', SVC(gamma='scale')))

    # lists to hold results of training
    resultsAccuracy = []
    names = []
    
    # loop through and train each model
    for name, model in models:
        model.fit([x.flatten() for x in x_train],np.argmax(y_train,axis=1)) # x,y are required to be 1-d lists not n*n arrays
        kfold = model_selection.KFold(n_splits=10, random_state=8623)
        accuracy_results = model_selection.cross_val_score(model, [x.flatten() for x in x_train],np.argmax(y_train,axis=1), cv=kfold, scoring='accuracy')
        resultsAccuracy.append(accuracy_results)
        names.append(name)
        accuracyMessage = "%s: %f (%f)" % (name, accuracy_results.mean(), accuracy_results.std())
        print(accuracyMessage)
   
    # boxplot algorithm comparison
    fig = plt.figure()
    fig.suptitle('Algorithm Comparison: Accuracy')
    ax = fig.add_subplot(111)
    plt.boxplot(resultsAccuracy)
    ax.set_xticklabels(names)
    ax.set_ylabel('Cross-Validation: Accuracy Score')
    plt.show()
    return
```

```python
TrainTest_ML_Models(x_train,y_train,x_test,y_test)
```

### Model 2 - Convoltional Neural Networks (CNN)
 - [a simple NN](https://victorzhou.com/blog/keras-neural-network-tutorial/)

#### Summary of CNN  Architecture
***conv2d layers***
- [convolutional matrices](https://beckernick.github.io/convolutions/) blend two or more data together. On images, this can be used to transform and extract feature from images such as [blurring or sharpenning](https://docs.gimp.org/2.8/en/plug-in-convmatrix.html). 
- *[filters](https://stackoverflow.com/a/48243703/4538066):* sets the dimensionality limit (*for RGB images, rule of thumb =* $x*y*3$)
- *kernel_size:* specifies height*width of convolutional window.
- *[padding](https://www.corvil.com/kb/what-is-the-difference-between-same-and-valid-padding-in-tf-nn-max-pool-of-tensorflow):* padding strategy at borders
- *input_shape:*** if first layer, specify ($x*y*z$) dimensions of the image.

***max pooling layers***
 - [pooling](https://machinelearningmastery.com/pooling-layers-for-convolutional-neural-networks/) is a similar concept to down sampling in statistics whereby small variations and noise are sampled out in order to preserve the primary structure and features of the data. In image analysis this has the additional benefit for the robust capturing of spatial features, such as lines and object boundaries, even if these boundaries vary slight between sequences of images (i.e. video).
 - *pool_size:* almost always 2×2 pixels applied with a stride of 2 pixels.

***Dense layers***
 - [dense layers](https://medium.com/@hunterheidenreich/understanding-keras-dense-layers-2abadff9b990) are you regular, fully connected nerual networks in which every input, hidden and output layer neuron is fully connected to neighbouring layers.
 
***activation functions***
 - activation functions turn arbitary real values to usefull scales and probabilities. 
 - *[ReLU](https://en.wikipedia.org/wiki/Rectifier_(neural_networks)#Advantages):* ramp function or recitifed linear units are simple, efficient, one-step.
 - *[softmax](https://victorzhou.com/blog/softmax/):* uses $e$ to convert arbitary real values to probabilities
 
***utilities***
 - *Dropout rate:* sets a fraction rate of input units to 0 at each update during training time, which helps prevent overfitting.
 - *flatten:* Flattens the input matrix to 1d vector.
 
 

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Activation
from keras.layers import Dropout
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras.optimizers import SGD
import keras.callbacks
```

```python
# network design
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', input_shape=(80, 80, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) # from 80*80 to 40x40
model.add(Dropout(0.25))

model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) # from 40x40 to 20x20
model.add(Dropout(0.25))

model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) # from 20x20 to 10x10
model.add(Dropout(0.25))

model.add(Conv2D(32, (10, 10), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2))) # from 10x10 to 5x5
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(2, activation='softmax'))
```

```python
num_epochs = 10

# optimization setup
sgd = SGD(lr=0.01, momentum=0.9, nesterov=True)

model.compile(
    loss='categorical_crossentropy',
    optimizer=sgd,
    metrics=['accuracy'])

# training
training_history = model.fit(X,
                             y,
                             batch_size=32,
                             epochs=num_epochs,
                             validation_split=0.2,
                             shuffle=True,
                             verbose=2)
```

```python
# plot trainng metrics
fig,axs = plt.subplots(1,4,figsize=(18,4))
for k,ax in zip(training_history.history.keys(),axs.flatten()):
    ax.plot(training_history.history[k])
    ax.set_title(k)
    ax.set_xticks(range(num_epochs))
    ax.set_xlabel('epoch')
```

```python
plt.imshow(confusion_matrix(np.argmax(model.predict(x_test),
                                      axis=1),np.argmax(y_test,axis=1)),
           interpolation='nearest',
           cmap=plt.cm.Blues);
```

```python
pd.crosstab(np.argmax(model.predict(x_test),axis=1),np.argmax(y_test,axis=1))
```
